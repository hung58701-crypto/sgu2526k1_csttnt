# ex_sum_even

## Nội dung bài
Nhập một số nguyên `n`.  
Hãy tính **tổng các số chẵn từ 1 đến n**.

## Bài làm

### Nhập dữ liệu

```python
n = int(input("Moi ban nhap so nguyen n: "))
```

### Xử lý

```python
s = 0
for i in range(1, n + 1):
    if i % 2 == 0:
        s += i
```

### Xuất dữ liệu

```python
print(f'Tong cac so chan tu 1 den {n} la {s}.')
```
