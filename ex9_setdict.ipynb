# ex9_set_dict

## Nội dung bài
Viết chương trình nhập vào một dãy ký tự `a` (các ký tự cách nhau bởi khoảng trắng).  
Thực hiện các yêu cầu sau:

1. In ra các ký tự **không trùng nhau** và được **sắp xếp tăng dần**
2. Đếm **số lần xuất hiện** của mỗi ký tự trong dãy `a`

---


## Bài làm

### Hàm xử lý các ký tự không trùng

```python
def XuLyKhongTrung(a):
    """
    Input:
        a = ['A', 'G', 'C', 'A', 'C', 'F', 'F', 'T', 'H']
    Output:
        s = ['A', 'C', 'F', 'G', 'H', 'T']
    """
    s = set([])
    for x in a:
        s.add(x)
        
    return sorted(s)
```

### Kiểm tra hàm XuLyKhongTrung

```python
a = ['A', 'G', 'C', 'A', 'C', 'F', 'F', 'T', 'H']
b = XuLyKhongTrung(a)
print(b)
```

### Hàm đếm số lần xuất hiện của ký tự

```python
def DemSoLanXuatHien(a):
    """
    Input:
        a = ['A', 'G', 'C', 'A', 'C', 'F', 'F', 'T', 'H']
    Output:
        dem = {'A': 2, 'G': 1, 'C': 2, 'F': 2, 'T': 1, 'H': 1}
    """
    dem = {}  # k: ky tu, v: so lan xuat hien
    for x in a:
        if x in dem:
            dem[x] += 1
        else:
            dem[x] = 1
    
    return dem
```

### Kiểm tra hàm DemSoLanXuatHien

```python
a = ['A', 'G', 'C', 'A', 'C', 'F', 'F', 'T', 'H']
b = DemSoLanXuatHien(a)
print(b)
```
