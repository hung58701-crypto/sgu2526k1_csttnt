
# ex_list_function

## Nội dung bài
Xây dựng các hàm xử lý **List** và kiểm tra tính đúng đắn của các hàm:

1. Nhập một mảng số từ bàn phím (các số cách nhau bởi khoảng trắng) và xuất mảng.
2. Sinh một mảng số ngẫu nhiên có `n` phần tử trong khoảng `[a, b]`.
3. Tính tổng các phần tử của mảng, đếm số phần tử **chẵn** và **lẻ**.
4. Cho một mảng số, trả về **dãy chẵn** và **dãy lẻ**.

---

## Bài làm



### Hàm nhập mảng


```python

def NhapMang():
    """
    Input:
        Ban phim = 1 4 1 2 3 4 1 2 3 3 1 2 3 4 10 2
    Output:
        a = [1, 4, 1, 2, 3, 4, 1, 2, 3, 3, 1, 2, 3, 4, 10, 2]
    """
    s = input("Nhap cac so (cach nhau boi khoang trang): ")
    a = list(map(int, s.split()))
    return a
```

### Hàm xuất mảng

```python

def XuatMang(a):
    """
    Input:
        a = [1, 4, 1, 2, 3, 4, 1, 2, 3, 3, 1, 2, 3, 4, 10, 2]
    Output: 
        Mang co 16 phan tu: 1 4 1 2 3 4 1 2 3 3 1 2 3 4 10 2
    """
    print(f'Mang co {len(a)} phan tu:', end=' ')
    for x in a:
        print(x, end=' ')
    print()
```

### Hàm sinh mảng ngẫu nhiên

```python

import random
def SinhNgauNhien(n, vmin = -10, vmax = 10):
    """
    Input:
        n = 30, vmin = -10, vmax = 10
    Output:
        a = [...]
    """    
    a = []
    for _ in range(n):
        a.append(random.randint(vmin, vmax))
    return a
```


### Hàm tính tổng, đếm chẵn – lẻ


```python
def DemTongChanLe(a):
    """
    Input: a[]
    Output: tong, sochan, sole
    """
    tong = 0
    sochan = 0
    sole = 0
    
    for x in a:
        tong += x
        if x % 2 == 0:
            sochan += 1
        else:
            sole += 1
    
    return tong, sochan, sole
```

### Hàm tách dãy chẵn – dãy lẻ

```python
def DayChanLe(a):
    """
    Input: a[]
    Output: (achan, ale)
    """    
    achan = []
    ale = []
    
    for x in a:
        if x % 2 == 0:
            achan.append(x)
        else:
            ale.append(x)
    
    return achan, ale
```

### Kiểm tra chương trình

```python
a = NhapMang()
XuatMang(a)

b = SinhNgauNhien(10, -10, 10)
XuatMang(b)

tong, chan, le = DemTongChanLe(b)
print(f'Tong = {tong}, So chan = {chan}, So le = {le}')

a_chan, a_le = DayChanLe(b)
print(f'Day chan: {a_chan}')
print(f'Day le: {a_le}')
```
