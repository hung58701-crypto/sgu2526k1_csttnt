# ex_hinh_chu_nhat
## Nội dung bài
Nhập vào chiều dài và chiều rộng của hình chữ nhật.  
Tính **chu vi** và **diện tích** của hình chữ nhật đó.

---


## Bài làm

---

### Nhập dữ liệu

```python
dai = float(input("Nhập chiều dài: "))
rong = float(input("Nhập chiều rộng: "))
```

### Xử lý tính toán

```python
chu_vi = 2 * (dai + rong)
dien_tich = dai * rong
```

### Xuất kết quả

```python
print(f"Chu vi hình chữ nhật là: {chu_vi}")
print(f"Diện tích hình chữ nhật là: {dien_tich}")
```
