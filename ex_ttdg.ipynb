## Bài làm

---


### Hiển thị tiêu đề

```python
print("Calculator Basic")
```

### Nhập dữ liệu

```python
a = int(input("Moi ban nhap so a: "))
b = int(input("Moi ban nhap so b: "))
```

### Xử lý tính toán

```python
c = a + b
```

### Xuất kết quả

```python
print("c = %d + %d = %d" % (a, b, c))
```
