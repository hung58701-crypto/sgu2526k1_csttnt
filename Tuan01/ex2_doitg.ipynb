# ex2_time_convert

## Nội dung bài
Một thiết bị hoạt động được $t$ giây.  
Hãy viết chương trình chuyển số giây đó dưới dạng **giờ : phút : giây**.

## Bài làm

```python
# Không cần thư viện ngoài
```

### Nhập dữ liệu

```python
t  = int(input('Nhap vao tong so giay: '))
```

### Xử lý

```python
hh = t // 3600
mm = (t % 3600) // 60
ss = t % 60
```

### Xuất dữ liệu

```python
print(f'{t} giay co dang {hh}:{mm}:{ss}')
```
