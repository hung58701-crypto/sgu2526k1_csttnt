# ex_prime_check

## Nội dung bài
Nhập vào một số nguyên dương `n`.  
Hãy kiểm tra xem `n` có phải là **số nguyên tố** hay không.


## Bài làm

### Nhập dữ liệu

```python
n = int(input("Moi ban nhap so nguyen n: "))
```

### Xử lý

```python
if n < 2:
    lant = False
else:
    lant = True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            lant = False
            break
```

### Xuất dữ liệu

```python
if lant == True:
    print(f'{n} la so nguyen to.')
else:
    print(f'{n} khong la so nguyen to.')
```
